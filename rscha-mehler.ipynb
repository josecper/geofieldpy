```python
def mehler_thebault(m, fn, theta, theta_0, norm=0):
    uf = numpy.frompyfunc(lambda x: mfunnorm.mfunnorm(m,fn, x, theta_0, norm)[0], 1, 1)
    return numpy.float64(uf(theta))
```
