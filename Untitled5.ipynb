```python
import numpy

import rscha2d
import fibonacci_sphere
import trig
import constants
import coords
import gauge
import geofield
import xyzfield

from matplotlib import pyplot, colors, ticker
import matplotlib
%matplotlib nbagg
matplotlib.rcParams["font.family"] = "Fira Sans"
matplotlib.rcParams.update({'font.size': 10})
```

```python
#lets do a thing to the data

fname = "../data/db_q_septiembre17.csv"
savepath = "../data/rscha2d/otros"
```

```python
lat_c = 42.0
lon_c = 20.0
theta_c, phi_c, theta_0d, theta_0 = numpy.deg2rad((90-lat_c, lon_c, 26.0, 50.0))
knots = numpy.arange(-2100, 1950, 50)
```

```python
sha = geofield.SHADIFData("../data/sha.dif.14k/coeff_SHA.DIF.14k.dat",
                          "../data/sha.dif.14k/error_coef_EPSL_tnmgh.dat")

base_years = numpy.arange(-2100, 1950, 50)
base_gcoefs = numpy.loadtxt("../bootstrap_javi/gh_mezcla_ord.dat").T[-len(base_years):, :]
```

```python
model = rscha2d.Model()
model.set_model_params(theta_c, phi_c, theta_0d,
                        cap_edge=0, kmax_int=2, kmax_ext=2, m_max=2,
                        knots=knots,
                        spatial_reg=1e-4, temporal_reg=1e-2,
                        refmodel={"years" : base_years,
                                  "coefs" : base_gcoefs})

model.add_data(fname)
```

```python
name_at = "Paris"
theta_at, phi_at = numpy.deg2rad((90-48.8, 2.35))

near = (trig.angulardist(theta_at, phi_at, model.thetav, model.phiv) 
        < numpy.arctan(900/constants.a_r))

D_reloc, I_reloc, F_reloc = trig.relocate(model.D_o, model.I_o, model.F_o,
                                          model.thetav, model.phiv, theta_at, phi_at)

t_cur = numpy.linspace(-2100, 1950, 400)
t_cur, r_cur, theta_cur, phi_cur = coords.curve_at(t_cur, constants.a_r, theta_at, phi_at)
```

```python
A = model.model_matrix(model.tw, model.r_geo,
                       model.theta_r, model.phi_r, 
                       reverse_order=True)[~model.nan_DIF]
data = numpy.concatenate((model.D_w,
                          model.I_w,
                          model.F_w/model.F_dip))[~model.nan_DIF]

M_reg = model.regularization_matrix()

n_knots = len(model.knots)
n_degrees = (len(model.k_int)
             + len(model.k_ext)
             + len(model.m_mehler))

m = numpy.zeros(A.shape[-1])
for i in range(5):
    D_syn, I_syn, F_syn = model.synth_data(model.tw,
                                           model.r_geo,
                                           model.theta_geo,
                                           model.phi_geo, m)
    F_syn = F_syn / model.F_dip
    data_syn = numpy.concatenate((D_syn, I_syn, F_syn))[~model.nan_DIF]
    m = m + numpy.linalg.lstsq(A.T @ A + M_reg, A.T @ (data - data_syn))[0]
    m = numpy.concatenate(m.reshape(n_degrees,n_knots).T)
    print(numpy.linalg.norm(data-data_syn))
```

```python
D_zero, I_zero, F_zero = model.synth_data(t_cur, r_cur, theta_cur, phi_cur, numpy.zeros(A.shape[-1]))
```

```python
D_test, I_test, F_test = model.synth_data(t_cur, r_cur, theta_cur, phi_cur, m)
```

```python
fig, (ax0, ax1, ax2) = pyplot.subplots(3,1)

ax0.plot(t_cur, D_zero)
ax1.plot(t_cur, I_zero)
ax2.plot(t_cur, F_zero)


ax0.plot(t_cur, D_test)
ax1.plot(t_cur, I_test)
ax2.plot(t_cur, F_test)
```
